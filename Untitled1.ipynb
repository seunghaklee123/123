```python
signal = SawtoothSignal(500)
wave = signal.make_wave(duration= 1, framerate=10000)
segment = wave.segment(duration = 0.005)
segment.plot()
decorate(xlabel='Time (s)')
```

주파수 500의 톱니파의 그래프를 ts = 10000으로 그리고 0.005까지 그래프를 그리기

```python
wave.make_audio()
```

위 신호를 소리로 만들어서 들어보기

```python
import numpy as np

hs = np.fft.rfft(wave.ys)
hs
```

wave에 해당하는 값을 rft해서 hs로 뽑기

```python
n=len(wave.ys)
d= 1/wave.framerate
fs = np.fft.rfftfreq(n,d)
fs
```

rft frequency에 n,d값을 넣어줘서 fs로 뽑기(주파수의 정수배)

```python
import matplotlib.pyplot as plt

maginitude = np.abs(hs)
plt.plot(fs, maginitude)
plt.xlabel('Frequency (Hz)')
```

hs에대한 magnitude를 뽑고 그 값과 fs를 그래프로 그리기 y값은 magnitude에서 나타남

```python
angle = np.angle(hs)
plt.plot(fs, angle)
plt.xlabel('phase(radian)')
```

hs에서 angle을 뽑고 그 값과 fs값으로 phase로 그래프 그리기
랜덤한 값이 나오는 것을 확인가능하다.

```python
import random
random.shuffle(angle)
plt.plot(fs, angle)
plt.xlabel('Phase(radian)')
```

```python
def view_harmonics(freq,framerate):
  """Plot the spectrum of a sawthooth signal.

  freq: frequency in Hz
  framerate: in frames/second
  """
  signal = SawtoothSignal(freq)
  wave = signal.make_wave(duration=0.5, framerate=framerate)
  spectrum = wave.make_spectrum()
  spectrum.plot(color='C0')
  decorate(xlabel='Frequency (Hz)',ylabel='Amplitude')
  display(wave.make_audio())
```

```python
from ipywidgets import interact,interactive,fixed
import ipywidgets as widgets

slider1 = widgets.FloatSlider(min =100, max=10000, value=100, step=100)
slider2 = widgets.FloatSlider(min =5000, max=40000, value=10000, step=1000)
interact(view_harmonics, freq = slider1,framerate = slider2);
```

파형의 주파수와 framerate를 변경하면서 aliasing여부를 소리와 그래프로 확인해보기
